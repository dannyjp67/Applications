```python
tags = {
    'public_transport':['station'],
}
```

```python
G = ox.graph_from_place('Greater London, UK', network_type='walk')
```

```python
G = ox.projection.project_graph(G, to_crs=27700)
```

```python
pois = ox.geometries.geometries_from_place('Greater London, UK', tags=tags)
```

```python
pois = pois.to_crs(epsg=27700)
centroids = pois.centroid
```

```python
walk_time = 20

walk_speed = 4.5
```

```python
for u, v, data in G.edges(data=True):
    data['speed_kph'] = walk_speed
G = ox.add_edge_travel_times(G)

# Extract node/edge GeoDataFrames, retaining only necessary columns (for pandana)
nodes = ox.graph_to_gdfs(G, edges=False)[['x', 'y']]
edges = ox.graph_to_gdfs(G, nodes=False).reset_index()[['u', 'v', 'travel_time']]
```

```python
attempt = ox.graph_to_gdfs(G, nodes=False)
```

```python
network = pandana.network.Network(
    node_x=nodes['x'],
    node_y=nodes['y'], 
    edge_from=edges['u'],
    edge_to=edges['v'],
    edge_weights=edges[['travel_time']]
)
```

```python
maxdist = walk_time * 60
```

```python
network.set_pois(
    category='pois',
    maxdist=maxdist,
    maxitems=10,
    x_col=centroids.x, 
    y_col=centroids.y
)
```

```python
distances = network.nearest_pois(
    distance=maxdist,
    category='pois',
    num_pois=2
)

distances.astype(int).head(10)
```

```python
COLOR = 'white'
plt.rcParams['text.color'] = COLOR
plt.rcParams['axes.labelcolor'] = COLOR
plt.rcParams['xtick.color'] = COLOR
plt.rcParams['ytick.color'] = COLOR

# Setup plot
fig, ax = plt.subplots(figsize=(20,15))
ax.set_axis_off()
ax.set_aspect('equal')
fig.set_facecolor((0,0,0))

# Plot distance to nearest POI
sc = ax.scatter(
    x=nodes['x'],
    y=nodes['y'], 
    c=distances[1],
    s=3,
    cmap='PuBu',
)

# Colorbar
cb = fig.colorbar(sc, ax=ax, shrink=0.8, ticks=[0, 300, 600, 900])
cb.ax.tick_params(color='none', labelsize=20)
cb.ax.set_yticklabels(['0', '5', '10', '>= 20'])
cb.set_label('Walking time to nearest Station (minutes)', fontsize=20, fontweight='bold')

# Remove empty space
plt.tight_layout()
```
