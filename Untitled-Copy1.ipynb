```python
lisa = esda.moran.Moran_Local(stapols['PM25'], wk)
```

```python
stapols['Is'] = lisa.Is
```

```python
stapols['p-sim'] = lisa.p_sim

sig = 1 * (lisa.p_sim < 0.05)
slabels = ['non-sig.', 'significant'] 
labels = [slabels[i] for i in sig]

stapols['sig'] = labels
stapols[['sig','p-sim']].head(10)
```

```python
stapols['sigIs'] = np.where(stapols['sig'] == 'significant', stapols['Is'], np.nan)
```

```python
counts = [(j,(lisa.q==j).sum()) for j in range(1,5)]
```

```python
qlabels = ['HH', 'LH', 'LL', 'HL']  #pysal scheme is  HH=1, LH=2, LL=3, HL=4
labels = [qlabels[i-1] for i in lisa.q]
```

```python
stapols['qlabels'] = labels
[(qlabel, (stapolsss['qlabels']==qlabel).sum()) for qlabel in qlabels]
```

```python
hotspot = 1 * (sig * lisa.q==1)
coldspot = 3 * (sig * lisa.q==3)
doughnut = 2 * (sig * lisa.q==2)
diamond = 4 * (sig * lisa.q==4)
spots = hotspot + coldspot + doughnut + diamond
spot_labels = [ ' non-sig.', ' hot spot', ' doughnut', ' cold spot', ' diamond']
labels = [spot_labels[i] for i in spots]
```

```python
stapols['slabels'] = labels
```

```python
fig, ax = plt.subplots(1, figsize=(9,9))

sigcolors = colors.ListedColormap([ 'blue', 'red','lightgrey'])

stapols.plot(column='slabels', categorical=True,
         k=3, cmap=sigcolors, linewidth=0.1, edgecolor='white', 
         legend=True, legend_kwds={"title":"LISA (Moran's I)","loc": 2},
         ax=ax)
plt.grid(False)
ax.set_axis_off()
plt.show()
```
